```python
#4 (1)
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.cm as cm  # colormap
import numpy as np
import math
plt.rcParams["text.usetex"] =True 
fig = plt.figure(figsize=(12,12))


ax1 = fig.add_subplot(111)

def plotmsd(moji,tau,v0,mgn,R,i):
    t,msd= np.loadtxt("C:/Users/ikedah/Documents/sotukenn/workspace/refwca/{}R{:.1f}v0{:.1f}tau{:.3f}m{:.3f}/msd_R{:.3f}_tau{:.3f}_m{:.3f}.dat".format(moji,R,v0,tau,mgn,R,tau,mgn), comments='#', unpack=True)
    plt.plot(t,msd, "-" ,markersize=5,color=cm.jet(i),label="tau{},mg{},R{}".format(tau,mgn,R))
    
plt.xscale('log')
plt.yscale('log')
#plt.xlim(0.01,60)
#plt.ylim(0.0001,10000)
plt.tick_params(which='major',width = 1, length = 10)
plt.tick_params(which='minor',width = 1, length = 5)
ax1.spines['top'].set_linewidth(3)
ax1.spines['bottom'].set_linewidth(3)
ax1.spines['left'].set_linewidth(3)
ax1.spines['right'].set_linewidth(3)
plt.xlabel(r"$t/t_0$",color='k', size=30)
plt.ylabel(r"$\langle \Delta \mathbf{r}(t)^2\rangle/a^2$",color='k', size=30)
plt.xticks(color='k', size=30)
plt.yticks(color='k', size=30)


#################################
plt.subplots_adjust(wspace=0.5, hspace=0.25)
moji="stwne4"
tau=10
mgn=0
R=28.868#22.361
tau=10
v0=1
i=0
plotmsd(moji,tau,v0,mgn,R,i)
# plotmsd(moji,tau,v0,0.01,R,2/5)
plotmsd(moji,tau,v0,0.1,R,3/5)
plotmsd(moji,tau,v0,1,R,4/5)
plotmsd(moji,tau,v0,2,R,5/5)
plt.legend(ncol=1, loc=4, borderaxespad=0, fontsize=20,frameon=False)
x=np.linspace(0.01,tau*2,100)
x2=np.linspace(tau,1000,500)
plt.plot(x,x*x,"--",color="b")
plt.plot(x2,x2,"--",color="g")

plt.savefig('C:/Users/ikedah/Documents/sotukenn/workspace/refwca/msdtau{}m{}R{}.png'.format(tau,mgn,R))
#plt.savefig(r'C:/Users/ikedah/Downloads/2022-simulation-training-main/2022-simulation-training-main/report/ene.pdf')
plt.show()
```
